```python
# dependencies
import pandas as pd
```

```python
# store filepath in a variable
zoomfile = "resources/zoominfo.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
zoom_org_df = pd.read_csv(zoomfile, encoding="ISO-8859-1")
zoom_df = zoom_org_df.copy()
```

```python
# show the header
zoom_org_df.head()
```

```python
# remove unnecessary columns
zoom_df = zoom_df.drop(['Date','Intent ID', 'Company ID','Category','Score','Recommended Contact ID'], axis=1)
zoom_df = zoom_df.drop(['Recommended Contact Notice Provided Date', 'Secondary Industry Hierarchical Category','Query Name'], axis=1)
zoom_df = zoom_df.drop(['Primary Industry', 'Primary Sub-Industry','All Industries','LinkedIn Company Profile URL'], axis=1)
```

```python
# remove rows that have inactive companies -'certified active company'
# certified active companies =='Yes'
index_names = zoom_df[zoom_df['Certified Active Company']== 'No'].index
zoom_df.drop(index_names, inplace = True)
zoom_df.head()
```

```python
# delete Certified Active Company column
zoom_df.drop(['Certified Active Company'], axis=1,inplace=True)
zoom_df.head()
```

```python
zoom_df.shape
```

```python
# split HQ phone column into two columns df.Name.str.split("_",expand=True)
phone_split = zoom_df['Recommended Contact Phone (Direct)'].str.split("e",expand=True)
phone_split
```

```python
# store split phone number in zoom_df
zoom_df['Contact- Direct Phone']=phone_split[0]
zoom_df['Contact- Direct Extension']=phone_split[1]
```

```python
# split Company Street Address column into two columns (adddress & suite no.)
address_split = zoom_df['Company Street Address'].str.split(",",expand=True)
address_split
```

```python
# store split address in zoom_df
zoom_df['Company- Address- Line 1']=address_split[0]
zoom_df['Company- Address- Line 2']=address_split[1]
```

```python
zoom_df.columns
```

```python
# delete columns that were split in two
# address column and phone column
zoom_df.drop(['Recommended Contact Phone (Direct)','Company Street Address',], axis=1,inplace=True)
# drop remaining unnecessary columns
zoom_df.drop(['All Sub-Industries','Industry Hierarchical Category',], axis=1,inplace=True)
```

```python
# rename current columns to connectwise format
zoom_df.rename(columns={'Topic': 'Company- Note',
                        'Recommended Contact First Name': 'Contact- First Name',
                        'Recommended Contact Last Name': 'Contact- Last Name',
                        'Recommended Contact Title':'Contact- Job Title',
                        'Recommended Contact Email':'Contact- Email address',
                        'LinkedIn Contact Profile URL':'Contact- Note',
                        'Website':'Company- Website URL',
                        'Company HQ Phone':'Company- Phone Number',
                        'Revenue (in 000s)':'Company- Annual Revenue',
                        'Employees':'Company- Number of employees',
                        'Company City':'Company- City',
                        'Company State':'Company- State',
                        'Company Zip Code':'Company- Zip',
                        'Company Country':'Company- Country'                     
                        },
          inplace=True, errors='raise')
```
